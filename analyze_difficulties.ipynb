```python
cat_dict = {}
for i in range(1024):
    niche = str(bin(i))[2:]
    zeros = 10 - len(niche)
    niche = '0' * zeros + niche
    cat_dict[niche] = []
```

```python
for puz_emb in embs:
    niche = ''.join([str(int(el)) for el in puz_emb['emb']])
    cat_dict[niche].append(puz_emb['program_str'])
```

```python
niche_and_nums = sorted([(k, len(v)) for k, v in cat_dict.items()], key=lambda x: x[1])[::-1]
```

```python
niche_and_nums
```

```python
def get_niche(program_str):
    program_strs = [p['program_str'] for p in all_solved_puzzles]
    if not program_str in program_strs:
        print('program_str not found!')
    else:
        index = program_strs.index(program_str)
        niche = ''.join([str(int(el)) for el in embs[index]['emb']])
        return niche
```

```python
get_niche(all_solved_puzzles[0]['program_str'])
```

```python
def get_count(program_str):
    niche = get_niche(program_str)
    return len(cat_dict[niche])
```

```python
get_count(all_solved_puzzles[0]['program_str'])
```

```python
for i, p in enumerate(puzzs_to_sample_from):
    print(f'puzzle {i}')
    print(get_count(p['program_str']))
    print('---')
```

```python
list_counts_niches_programs = [(p['program_str'], get_count(p['program_str']), get_niche(p['program_str'])) for p in puzzs_to_sample_from]
```

```python
# remove niche 0
list_counts_niches_programs = [lcn for lcn in list_counts_niches_programs if lcn[2] != "0" * 10]
len(list_counts_niches_programs)
```

```python
lcn_sorted = sorted(list_counts_niches_programs, key=lambda x: x[1])[::-1]
# puzzs_to_use = [lcn[0] for lcn in lcn_sorted][:3]  # the three puzzles in the most populated niche, duplicate niches allowed
puzzs_to_use = []
niches_used = []
idx = 0
while len(puzzs_to_use) != 4 and idx < len(lcn_sorted):
    program, count, niche = lcn_sorted[idx]
    if niche not in niches_used:
        puzzs_to_use.append(program)
        niches_used.append(niche)
    idx += 1
```

```python
for p in puzzs_to_use:
    print(p)
    print('\n' + "=" * 80 + '\n')
```
