```python
import ast
```

```python
def parse_puzzles(s):
    parsed = ast.parse(s)
    funs = [el for el in parsed.body if isinstance(el, ast.FunctionDef)]
    f = funs[0]
    g = funs[1]
    sat = ast.unparse(f)
    sol = ast.unparse(g)
    sol_header = sol.split(':\n')[0] + ':'
    sol_body = '\n'.join(sol.split(':\n')[1:])
    return sat, sol_header, [sol_body]
```
